```python
import pandas as pd
import numpy as np
#Read 2019 Player Stats per 36 minutes into a DataFrame
NBA=pd.read_csv('Desktop/python/2019PlayerStats.csv')

#Display first 5 rows of NBA
NBA.head()
```

```python
#Notice that Player name concatenates some sort of player ID.

#Replace \ with / to avoid escape character
NBA["Player"]=NBA["Player"].str.replace("\\","/")
NBA.head()
```

```python
#Chop off ID part.

NBA["Player"]=NBA["Player"].str.split("/").str[0]
NBA.head()
```

```python
#Pull all shooting guards
sgs=NBA[NBA["Pos"]=="SG"]
NBA[NBA["Pos"]=="SG"].head()
```

```python
#Show top 5 players in Points/36
ppg=NBA.sort_values(by=["PTS"],ascending=False)

ppg.head()

#Zhou Qi: the most dominant scorer in NBA history
```

```python
#filter out players with under 500 minutes played
NBA=NBA[NBA["MP"]>500]

#Rerun PPG
ppg=NBA.sort_values(by=["PTS"],ascending=False)
ppg=ppg[["Player","PTS"]]
ppg.head()
```

```python
#import contract data

contracts=pd.read_csv('Desktop/python/NBAcontracts.csv')

#same edit to Player column
contracts["Player"]=contracts["Player"].str.split("\\").str[0]
contracts.head()
```

```python
#join ppg and contracts DataFrames

#ppg.join(contracts,on="Player")

ppg_contracts=pd.merge(ppg,contracts,on="Player")
ppg_contracts=ppg_contracts[["Player","PTS","2018-19"]]

for x in range(0,len(ppg_contracts)):
    ppg_contracts['2018-19'].replace('$',"")


ppg_contracts.head()
```

```python
#change contract string to int
for x in range(0,len(ppg_contracts)):
    a=ppg_contracts['2018-19'][x][1::].rstrip(' ').replace(',','')
    ppg_contracts['2018-19'][x]=int(a)

ppg_contracts.head()
```

```python
#Change PTS column to int to run stats on it

#TOO SLOW

for x in range(0,len(ppg_contracts)):
    ppg_contracts['PTS'][x]=int(ppg_contracts['PTS'][x])
```
